```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


country_by_1960_t = country_by_1960.set_index('Country Name')


non_year_columns = ['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68']
country_by_1960_t = country_by_1960_t.drop(columns=non_year_columns)


country_by_1960_t.columns = country_by_1960_t.columns.astype(str)
country_by_1960_t = country_by_1960_t.apply(pd.to_numeric, errors='coerce')


for country_name, data_values in country_by_1960_t.iterrows():
    
    print(f"Processing country: {country_name}")
    print("Data index (years):", data_values.index)
    print("Data values:", data_values.values)
    print("Data index types:", data_values.index.dtype)
    print("Data values types:", data_values.values.dtype)

    
    fig = plt.figure(figsize=(10, 5))
    
    
    sns.barplot(x=data_values.index, y=data_values.values)
    

    plt.xlabel('Years')
    plt.ylabel('Data Values')
    plt.title(f"{country_name} - Data Values from 1960 to 2022")
    
    
    plt.xticks(rotation=90)
    plt.show()
```
